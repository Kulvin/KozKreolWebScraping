```python
# import library
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_rows', None)
pd.set_option('display.max_colwidth', None)

import json




t={}
dataF=[]


web = requests.get(
    'https://www.lalitmauritius.org/en/dictionary.html?letter=d')

data = web.content
soup = BeautifulSoup(data, features="html.parser")
dicList = soup.find("ul", {"id":"dictionarylist"})



li_tags = dicList.findAll('li')

# s = str(dicList)

# s_replace = s.replace('<br/>', "\n")

# # while True:
# #     index_begin = s_replace.find("<")
# #     index_end = s_replace.find(">",index_begin + 1)
# #     if index_begin == -1:
# #         break
# #     s_replace = s_replace.replace(s_replace[index_begin:index_end+1],"")

# print(s_replace)

c=0
for word in li_tags:
    try:
        kreol = word.find("span",{"class":"main"}).text.replace('\n',"")
        
    except:
        kreol = None
        
    try:
        eng = word.find("div",attrs = {'class':'desc'})
        s = str(eng)
        s_replace = s.replace('<br/>', "\n")
        
        while True:
            index_begin = s_replace.find("<")
            index_end = s_replace.find(">",index_begin + 1)
            if index_begin == -1:
                break
            s_replace = s_replace.replace(s_replace[index_begin:index_end+1],"")

    
    except:
        eng = None
        

           
    dictionary = {"kreol":kreol,"eng":s_replace}
    c=c+1
    dataF.append(dictionary)
    
df = pd.DataFrame(dataF)

# df.to_json(r'\Users\Naweed\Desktop\test.json')
df.to_csv(r'\Users\Naweed\Desktop\test2.csv',index = False)


df.head(500)
# print(df)
```
