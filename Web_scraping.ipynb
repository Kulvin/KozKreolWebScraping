```python
# import library
import requests
from bs4 import BeautifulSoup
import pandas as pd
pd.set_option('display.max_colwidth', None)
import json


t={}
dataF=[]


web = requests.get(
    'https://www.lalitmauritius.org/en/dictionary.html?letter=a')

data = web.content
soup = BeautifulSoup(data, features="html.parser")
dicList = soup.find("ul", {"id":"dictionarylist"})



li_tags = dicList.findAll('li')

c=0
for word in li_tags:
    try:
        kreol = word.find("span",{"class":"main"}).text.replace('\n',"")
    except:
        kreol = None
        
    try:
        eng = word.find("div",{"class":"desc"}).text.replace('\n',"")
    except:
        eng = None
   
    dictionary = {"kreol":kreol,"eng":eng}
    c=c+1
    dataF.append(dictionary)
    
df = pd.DataFrame(dataF)
df.to_json(r'\Users\Naweed\Desktop\test.json')
df.to_csv(r'\Users\Naweed\Desktop\test.csv',index = False)


df.head(500)
 


```
